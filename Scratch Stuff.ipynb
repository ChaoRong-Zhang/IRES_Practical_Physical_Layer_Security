```python
mwp = [0,1,2,3,4]
dbs = [sub_df1, sub_df2, sub_df3, sub_df4]
nra_max = [100]
nra_min = [100]
nra_avg = [100]
for db in dbs :
    nra_max.append(np.max(db['M.N.P.R']))
    nra_min.append(np.min(db['M.N.P.R']))
    nra_avg.append(np.mean(db['M.N.P.R']))
#nra_max[0] = 5
plt.plot(mwp,nra_max,'-k',label='MAX')
plt.plot(mwp,nra_min,'-k',label='MIN')
plt.plot(mwp,nra_avg,'-b',label='AVG')
plt.title('Number of Packets Received By Eve')
plt.xlabel('Maximum number of "Weird Patterns"')
plt.xticks(mwp,mwp)
plt.legend()
plt.show()
```
